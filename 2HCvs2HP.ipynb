# Functions

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import cm 
%matplotlib inline


def plot_figure(dataset1,dataset2,name,colour,lw,ax):
    ax.scatter(x,dataset1,marker='_',c=colour, s=300,linewidth=1)
    ax.scatter(x,dataset2,marker='_',c=colour, s=300,linewidth=1)
    ax.plot(x,dataset1,linestyle=':',color=colour,linewidth=lw, label=name)
    ax.plot(x,dataset2,linestyle=':',color=colour,linewidth=lw)
    
    
def plot_style_A():
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['left'].set_linewidth(1.5)
    ax.spines['bottom'].set_visible(False)
    plt.tick_params(axis='x', top='off')
    ax.yaxis.set_ticks_position('left')
    ax.xaxis.set_tick_params(labelsize=12,width=1.5)
    ax.yaxis.set_tick_params(labelsize=12,width=1.5)
    plt.ylabel('Relative Energy (eV)',fontsize=12)
    plt.tight_layout()

def xlabels(labelnames):
    labels = [item.get_text() for item in ax.get_xticklabels()]
    labels = [i for i in labelnames]
    plt.xticks(x, labels,rotation=30,fontsize=10)
def legend_style_horizontal():
    box = ax.get_position()
    ax.set_position([box.x0, box.y0 + box.height * 0.05,
                 box.width, box.height * 0.95])

    ax.legend(loc='upper center', bbox_to_anchor=(0.2, +1.),
            ncol=2,fontsize=10,markerscale=0.75, frameon=False)

def legend_style_vertical():
    box = ax.get_position()
    ax.set_position([box.x0, box.y0 + box.width*0.05,
                 box.width*0.95, box.height])

    ax.legend(loc='best', bbox_to_anchor=(0.2, 1),
            ncol=1,fontsize=10,markerscale=0.75, frameon=False)
```

# Experimental observations

Publications:

2HP: Zhang et. al. *Chem. Commun.*, **2016**, 52, 6577-6580<br />
2HC: Zhang et. al. *Ang. Chem. Int.Ed.*, **2015**, 54, 8369–8373

**2HP**:

* Absorption = 357 nm (3.47 eV) in DCM
* Faint emission in DCM, not picked up by fluorescence spectrometry
* Frozen samples (77 K) show strong fluorescence (483–524 nm) assigned to keto emission. No enol emission
* Substituent effects:
 * **1** shows a green fluorescence peak at 512 nm (2.42 eV)
 * para subst has blue shift wrt **1**: **2** and **3** = 487 nm (2.55 eV) and 483 nm (2.57)
 * meta subst has red shift wrt **1**: **4** emission = 524 nm (2.37 eV)
* Packing modes:
 * **1** and **4**:Herringbone edge-face packing, no $\pi$ stacking
 * **2**: Supramolecular network, face-face packing but no $\pi$ interactions
 * **3**: Face-Face, antiparallel stacked 

# Analysis

## Vacuum

First we analyse the energies in vacuum of 2HP and 2HC molecules:

### 2HP

```python
fig,ax=plt.subplots(figsize=(12,6))
colours=['red','blue','green','magenta','black']

x=range(len(hp1_s0))

for i in range(len(hp_s0[:4])):
    
    plot_figure(hp_s0[i],hp_s1[i],'HP'+str(i+1),colours[i],1.5,ax)
x=[0,1,2,3,4]
plot_figure(hp_s0[4],hp_s1[4],'HP'+str(4+1),colours[4],1.5,ax)
labelnames=["E*CI","E*min","FC","K*min","K*MECI"]
xlabels(labelnames)
ax.xaxis.set_tick_params(labelsize=12,width=1.5)
ax.yaxis.set_tick_params(labelsize=12,width=1.5)
plt.ylabel('Relative Energy (eV)',fontsize=12)
ax.legend(loc='upper center', bbox_to_anchor=(0.8, +0.9),
            ncol=2,fontsize=12,markerscale=0.2, frameon=False)  
```

Now we have a good picture of the mechanism for each 2HP compound. Lets deconstruct the process with a view to the effect of the substituents. 


#### Vertical Excitations

```python
hp_VE=[i[2] for i in hp_s1]
print(hp_VE)
print("Mean: {} eV\nStandard deviation {}".format(np.mean(hp_VE).round(3),np.std(hp_VE).round(3)))
```

```python
fig,ax=plt.subplots()
ax.bar(range(1,6),hp_VE)
plt.xlabel("Compound")
plt.ylabel("Energy (eV)")
plt.ylim(3,4.5)
```

A quick reminder of the substituents:
1. H
2. m-F
3. m-OCH3
4. p-F
5. p-OCH3

Evidently, substitution at meta-position destabilises the S${_1}$ vertical excitation energy, whilst the para position stablisses, with respect to compound **1**. Methoxy in para stablises the most (**3**). These excitations are all to S1, bright, and HOMO-LUMO in character. The difference density plots show that density is lost from the phenol right oxygen and donated to the C=O carboynl. in 2 and three, there is some donation from the amino N. In 4, electron density is lost from the F. **5** involves large donation from the methoxy group and large antibonding character in the C=C double bond, explaining the stability of the Enol ZtoE minumum. 

#### Deactivation Pathways

Relaxation in S${_1}$ from the FC point results in ESIPT for all five systems. This can be rationalised by the increased basicity of the carbonyl oxygen, and the acidity of the proton by the loss of electron density from the phenol oxygen. ESIPT is followed by intramolecular rotation of the phenoxy ring, resulting in a large stablisation of the KS${_1}$ state. Stablisation energy wrt FC VE:

```python
for i in range(5):
    print("Compound {}: K*min - VE = {} eV".format(i+1,(hp_s1[i][3]-hp_VE[i]).round(3)))
```

The KS${_1}$ conical intersection is then almost degenerate with the Kminimum. 

So, ESIPT will be the heavily-favoured mechanism for each of compounds **1**-**5**. However, we were able to locate several minima in the Enol channel. One pathway involves intramolecular rotation to the trans-Enol form, whilst there also exist minima and approx. 30 degrees rotation in the cis-Enol form. However, the lowest energy minimum for each compound involves pyramidalisation of the unsaturated chain of the molecule, in an Z to E 90degree twist. The LUMO orbital becomes antibonding across this bond, rationalising this behaviour. However, this mechanism will almost certainly be secondary to the ESIPT process. A rudimentary scan of this rotation at the 3-21(d) level shows a barrier of at least 0.2 eV for this mode. There is an associated conical intersection in this channel for all except **3**, which couldnt be converged (final gap = 0.03 eV)

### 2HC

```python
fig,ax=plt.subplots(figsize=(12,6))
colours=['red','blue','green','magenta','black']

x=range(len(hp1_s0))
labelnames=["E*CI","E*min","FC","K*min","K*MECI"]
xlabels(labelnames)
for i in range(len(hc_s0[:3])):
    
    plot_figure(hc_s0[i],hc_s1[i],'HC'+str(i+1),colours[i],1.5,ax)
x=[0,2,3,4]
plot_figure(hc_s0[3],hc_s1[3],'HC'+str(3+1),colours[3],1.5,ax)
plot_figure(hc_s0[4],hc_s1[4],'HC'+str(4+1),colours[4],1.5,ax)

ax.xaxis.set_tick_params(labelsize=12,width=1.5)
ax.yaxis.set_tick_params(labelsize=12,width=1.5)
plt.ylabel('Relative Energy (eV)',fontsize=12)
ax.legend(loc='upper center', bbox_to_anchor=(0.4, +0.8),
            ncol=2,fontsize=12,markerscale=0.2, frameon=False) 
```

For 2HC, a qualitatively similar mechanistic picture arises albeit with some key differences. The substituent position has a larger effect on the bias of the relaxation channel due to the increased electron density perturbation in the excited state. For **4** and **5**, there is no Enol minimum on the PES, as ESIPT is preferred upon optimisation. Conical intersections in the K channel are extremely stable. The anomoly in the K channel is for **5**, where the angle of rotation to the Kminimum is only 23degrees. The conical intersection seam in **5** is also approached at approx 70degrees of rotation (see scan of dihedral angle, cannot converge after this point)

#### Vertical excitations

```python
hc_VE=[i[2] for i in hc_s1[:3]]
hc_VE.append(hc4_s1[1])
hc_VE.append(hc5_s1[1])
print(hc_VE)
print("Mean: {} eV\nStandard deviation {}".format(np.mean(hc_VE[:3]).round(3),np.std(hc_VE[:3]).round(3)))
```

The vertical excitation energies are insensitive to the substituent unless there is a strong donor in the para position, which stablises S${_1}$. This is also observed in 2HP-5.
